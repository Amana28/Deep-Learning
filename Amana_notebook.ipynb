```python
from PIL import Image

# Import an image from directory:
input_image = Image.open("cat.jpg")

# Resizing image to be smaller

new_image = input_image.resize((320, 319))
# new_image.save('smaller_cat.jpg')

# Extracting pixel map:
pixel_map = new_image.load()

# Extracting the width and height
# of the image:
width, height = new_image.size

print(width)
print(height)

# Changing top right 30X30 pixels to black
for i in range(width - 29, width):
	for j in range(0, 31):
		
		# getting the RGB pixel value.
		r, g, b = new_image.getpixel((i, j))
		
		# Apply formula of grayscale:
		grayscale = (0*r + 0*g + 0*b)

		# setting the pixel value.
		pixel_map[i, j] = (int(grayscale), int(grayscale), int(grayscale))

new_image
```
